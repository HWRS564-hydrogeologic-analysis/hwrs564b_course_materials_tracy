```python
# Organic contaminant phase partitioning
# Based on my Unit 5 slide #14 "Air-soil-water partitioning of organic contaminants":
#   C_air = H * C_water
#   Kd = Cs / Cw
#   Kd = foc * Koc

#Given in problem
compounds = {
    "Benzene": {
        "Cw_mg_L": 5.0,
        "H": 0.22,
        "Koc_L_kg": 83
    },
    "TCE": {
        "Cw_mg_L": 2.0,
        "H": 0.40,
        "Koc_L_kg": 126
    }
}

#Given in problem
foc = 0.02          # fraction organic carbon
rho_b = 1.6         # kg/L bulk density

def phase_distribution(name, Cw, H, Koc, foc, rho_b):
    # 1) Soil-water partition coefficient
    Kd = foc * Koc  # L/kg

    # 2) Sorbed concentration on solids
    Cs_mg_per_kg = Kd * Cw  # mg/kg

    # 3) Convert sorbed phase to mg/L bulk medium
    Csoil_mg_per_L = Cs_mg_per_kg * rho_b  # mg/L

    # 4) Air concentration from Henry's law
    Cair = H * Cw  # mg/L

    # 5) Total across these three phases
    Ctotal = Cw + Cair + Csoil_mg_per_L

    # 6) Fractions
    fw = Cw / Ctotal
    fair = Cair / Ctotal
    fsoil = Csoil_mg_per_L / Ctotal

    return {
        "compound": name,
        "Kd_L_kg": Kd,
        "Cs_mg_per_kg": Cs_mg_per_kg,
        "Csoil_mg_per_L": Csoil_mg_per_L,
        "Cair_mg_per_L": Cair,
        "Ctotal_mg_per_L": Ctotal,
        "fraction_water": fw,
        "fraction_air": fair,
        "fraction_soil": fsoil
    }

results = []
for name, vals in compounds.items():
    results.append(
        phase_distribution(
            name=name,
            Cw=vals["Cw_mg_L"],
            H=vals["H"],
            Koc=vals["Koc_L_kg"],
            foc=foc,
            rho_b=rho_b
        )
    )

for r in results:
    print(f"\n--- {r['compound']} ---")
    print(f"Kd = {r['Kd_L_kg']:.3f} L/kg")
    print(f"Cs = {r['Cs_mg_per_kg']:.3f} mg/kg")
    print(f"Csoil = {r['Csoil_mg_per_L']:.3f} mg/L")
    print(f"Cair = {r['Cair_mg_per_L']:.3f} mg/L")
    print(f"Ctotal = {r['Ctotal_mg_per_L']:.3f} mg/L")
    print(f"Water fraction = {r['fraction_water']:.3%}")
    print(f"Air fraction   = {r['fraction_air']:.3%}")
    print(f"Soil fraction  = {r['fraction_soil']:.3%}")
```
