```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the clean sheet
df = pd.read_excel(
    "ChemHw4_Langston.xlsx",
    sheet_name="Q4_clean",
    parse_dates=["Date"]
)

# Create figure with three stacked plots
fig, ax = plt.subplots(3, 1, figsize=(8,10), sharex=True)

# Plot DIN concentration
ax[0].plot(df["Date"], df["DIN [mg N L-1]"], marker="o")
ax[0].set_ylabel("DIN (mg N L⁻¹)")
ax[0].set_title("DIN Concentration")

# Plot event DIN leaching
ax[1].plot(df["Date"], df["Event DIN Leaching [mg N m-2]"], marker="o")
ax[1].set_ylabel("Event DIN (mg N m⁻²)")
ax[1].set_title("Event DIN Leaching")

# Plot accumulated DIN leaching
ax[2].plot(df["Date"], df["Accum DIN Leaching  [mg N m-2]"], marker="o")
ax[2].set_ylabel("Accumulated DIN (mg N m⁻²)")
ax[2].set_title("Accumulated DIN Leaching")

# Format x-axis
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load data
df = pd.read_excel(
    "ChemHw4_Langston.xlsx",
    sheet_name="Q4_clean",
    parse_dates=["Date"]
)

fig, ax1 = plt.subplots(figsize=(10,5))

# DIN (left axis)
line1 = ax1.plot(
    df["Date"],
    df["DIN [mg N L-1]"],
    color="blue",
    marker="o",
    label="DIN concentration"
)

ax1.set_ylabel("DIN (mg N L⁻¹)", color="blue")

# Right axis
ax2 = ax1.twinx()

line2 = ax2.plot(
    df["Date"],
    df["Event DIN Leaching [mg N m-2]"],
    color="red",
    marker="o",
    label="Event DIN leaching"
)

line3 = ax2.plot(
    df["Date"],
    df["Accum DIN Leaching  [mg N m-2]"],
    color="green",
    marker="o",
    label="Accumulated DIN leaching"
)

ax2.set_ylabel("DIN Leaching (mg N m⁻²)")

# ---- Improved date formatting ----

ax1.xaxis.set_major_locator(mdates.MonthLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%b-%y"))

# Force axis to start in Dec 2017
ax1.set_xlim(pd.Timestamp("2017-12-01"), pd.Timestamp("2018-07-01"))

# Grid
ax1.grid(True, linestyle="--", alpha=0.6)

# Legend
lines = line1 + line2 + line3
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc="upper left")

plt.title("DIN Concentration and Leaching (Dec 2017 – Jun 2018)")
plt.xlabel("Sampling Date")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("ChemHw4_Langston.xlsx", sheet_name="Q4_clean", parse_dates=["Date"])

plt.figure(figsize=(9,5))
plt.plot(df["Date"], df["DOC [mg C L-1)"], marker="o", color="purple", label="DOC")
plt.plot(df["Date"], df["Net_N [mg N L-1 day-1]"], marker="o", color="orange", label="Net microbial N production")
plt.plot(df["Date"], df["DIN [mg N L-1]"], marker="o", color="blue", label="DIN")

import matplotlib.dates as mdates

ax = plt.gca()

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b-%y"))

ax.set_xlim(pd.Timestamp("2017-12-01"), pd.Timestamp("2018-07-01"))

plt.ylabel("Concentration / Rate")
plt.title("DOC, Net Microbial N Production, and DIN Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
